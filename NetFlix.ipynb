```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
nf = pd.read_csv('netflix_titles.csv')
nf.shape
```

```python
nf.info()
nf.head()
```

```python
nf.isnull().sum()
```

```python
nf.dropna(subset=['director', 'cast'], inplace=True)
print('Before Filling')
print(nf)
nf['country'].fillna('Unknown', inplace=True)
nf['date_added'].fillna('Unknown', inplace=True)
nf['rating'].fillna(nf['rating'].mode()[0], inplace=True)
print('\nAfter Filling')
print(nf)
```

```python
print(nf)
```

```python
nf.describe(include='all')
```

```python
plt.figure(figsize=(10,4))
sns.histplot(nf['release_year'],kde=True,color='g')
plt.title('Distribution of Release Year')
plt.show()
```

```python
movies_df = nf[nf['type'] == 'Movie'].copy()
movies_df['duration'] = movies_df['duration'].str.replace(' min', '', regex=False).astype(float)

plt.figure(figsize=(8, 5))
sns.boxplot(y=movies_df['duration'])
plt.title('Movie Duration (in minutes)')
plt.show()

top_countries = nf['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries on Netflix')
plt.xlabel('Number of Titles')
plt.show()

numeric_df = nf.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
nf['duration'] = nf['duration'].astype(str).str.extract(r'(\d+)').astype(float)
plt.figure(figsize=(8, 5))
sns.boxplot(data=nf[nf['type'] == 'Movie'], x='duration')
plt.title('Boxplot of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.show()
```

```python
## Key Insights:

- Most Netflix content was released after 2015.
- United States has the highest number of titles.
- TV Shows are generally shorter per episode, while Movies vary in duration.
- Certain genres like Documentaries and Dramas are more frequent.
- Movie durations have outliers, with a few very long titles.
```
